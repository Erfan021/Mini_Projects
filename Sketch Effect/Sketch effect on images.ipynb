Converting to Grayscale

###### Sketch Effect

```python
img1_cartoon= cv2.stylization(img_1, sigma_s=500, sigma_r=0.1)
#the more the sigma_s the more smooth image
#the less the sigma_r the smooth the image
```

```python
cv2.imshow('cartoon', img1_cartoon)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img1_sketch, img_sketch2= cv2.pencilSketch(img_1, sigma_s=60, sigma_r=0.1)
```

```python
cv2.imshow('sketch', img1_sketch)
cv2.waitKey()
cv2.destroyAllWindows()
```
